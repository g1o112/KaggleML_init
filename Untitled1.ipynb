```python
import pandas as pd

melbourn_file_path = 'melb_data.csv'
melbourn_data = pd.read_csv(melbourn_file_path)

melbourn_data
```

```python
melbourn_data = melbourn_data.dropna(axis=0)
melbourn_data
```

```python
y = melbourn_data.Price # selecionando a coluna Price dentro do data para ser o y
#as colunas que vamos colocar no nosso modelo para mais tarde fazer previsoes com elas, são chamadas de >>>>>>features<<<<
# que no nosso caso essa scolunas vao ser usadas para o sistema determinar o valor da casa.
```

```python
# Each item in that list should be a string (with quotes).
melbourn_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = melbourn_data[melbourn_features] #não pode ser direto pois ele procura dentro do data oficial as variaceis de mesmo nome
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

#Definir modelo. Especifique um numero para random_state para garantir o mesmo resultado em todas as vezes
melbourn_model = DecisionTreeRegressor(random_state=1)

#fit model
melbourn_model.fit(x,y)
```

```python
#semrpe garantir o mesmo resultados em cada run
# agora temos um modelo definido para usar e fazer previsoes

print("Fazendo previsoes com as primeiras 5 casas:")
print(x.head())
print("A previsao é:")
print(melbourn_model.predict(x))
```

```python
from sklearn.metrics import mean_absolute_error

predi_home_p = melbourn_model.predict(x)
mean_absolute_error(y, predi_home_p) #in-sample o erro
```
