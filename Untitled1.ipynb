```python
import pandas as pd

melbourn_file_path = 'melb_data.csv'
melbourn_data = pd.read_csv(melbourn_file_path)

melbourn_data
```

```python
melbourn_data = melbourn_data.dropna(axis=0)
melbourn_data
```

```python
y = melbourn_data.Price # selecionando a coluna Price dentro do data para ser o y
#as colunas que vamos colocar no nosso modelo para mais tarde fazer previsoes com elas, são chamadas de >>>>>>features<<<<
# que no nosso caso essa scolunas vao ser usadas para o sistema determinar o valor da casa.
```

```python
# Each item in that list should be a string (with quotes).
melbourn_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = melbourn_data[melbourn_features] #não pode ser direto pois ele procura dentro do data oficial as variaceis de mesmo nome
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

#Definir modelo. Especifique um numero para random_state para garantir a consistencia e o mesmo resultado em todas as execuções
melbourn_model = DecisionTreeRegressor(random_state = 1)
forest_model = RandomForestRegressor(random_state = 1)

#fit model
melbourn_model.fit(x,y)
forest_model.fit(x,y)

print(melbourn_model)
print(forest_model)
```

```python
#semrpe garantir o mesmo resultados em cada run
# agora temos um modelo definido para usar e fazer previsoes

print("Fazendo previsoes com as primeiras 5 casas:")
print(x.head())
print("A previsao é:")
print(melbourn_model.predict(x))
print("A previsao forest é:")
print(forest_model.predict(x))
```

```python
from sklearn.metrics import mean_absolute_error

predi_home_p = melbourn_model.predict(x)
predi_home_forest = forest_model.predict(x)
print(mean_absolute_error(y, predi_home_p)) #in-sample o erro
print(mean_absolute_error(y, predi_home_forest)) #in-sample o erro
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_test, y_treino, y_test = train_test_split(x, y, random_state = 0)

melbourn_model = DecisionTreeRegressor()

melbourn_model.fit(x_treino, y_treino)
```

```python
val_predction = melbourn_model.predict(x_test)
print(mean_absolute_error(y_test, val_predction))
```
